```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('sale.csv')
data.head(3)
```

```python
X=np.array(data.iloc[:,1:4])
```

```python
X
```

```python
Y=np.array(data.iloc[:,-1])
```

```python
Y
```

```python
def calculate_Y_predict(w,X):
    return np.sum(w*X,axis=1)
```

```python
def cost(w,X):
    m,n=X.shape
    return sum((Y-calculate_Y_predict(w,X))**2)/(2*m)
```

```python
def grad(w,X):
    gr=[]
    m,n= X.shape
    for i in range(n):
        value=sum((calculate_Y_predict(w,X)-Y)*X[:,i])/m
        gr.append(value)
    return gr
```

```python
m,n= X.shape
error_rate=[]
w=[0 for i in range(n)]
epoch =2000
lr = 0.000001
for i in range(epoch):
    print('iter= {}  ||   mse={}   '.format(i,cost(w,X)))
    error_rate.append(cost(w,X))
    gr=grad(w,X)
    w=[w[i]- lr*gr[i] for i in range(len(gr))]
    print(w)
    print('=====================================================')
```

```python
iterator=[i for i in range(0,len(error_rate))]
```

```python
plt.plot(iterator,error_rate)
plt.title('Error Rate')
plt.xlabel('Training Iterations')
plt.ylabel('Mean Squared Error')
plt.axis([0,1000,0,120])
plt.show()
```
