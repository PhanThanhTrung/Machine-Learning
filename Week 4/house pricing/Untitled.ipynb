```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data= pd.read_csv('train.csv')
data.head(3)
```

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
corr=data.corr()
corr.style.background_gradient(cmap='coolwarm')
```

```python
data.OverallQual.isnull().sum()
data.YearBuilt.isnull().sum()
data.TotalBsmtSF.isnull().sum()
data.GrLivArea.isnull().sum()
data.FullBath.isnull().sum()
data.GarageCars.isnull().sum()
```

```python
x1=np.array([data.OverallQual])
x2=np.array([data.YearBuilt])
x3=np.array([data.TotalBsmtSF])
x4=np.array([data.GrLivArea])
x5=np.array([data.FullBath])
x6=np.array([data.GarageCars])
```

```python
x1.shape
```

```python
X=np.concatenate((x1,x2), axis=0)
X=np.concatenate((X,x3), axis=0)
X=np.concatenate((X,x4), axis=0)
X=np.concatenate((X,x5), axis=0)
X=np.concatenate((X,x6), axis=0)
```

```python
Y=np.array([data.SalePrice])
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model=model.fit(X.T,Y.T)
```

```python
data=pd.read_csv('test.csv')
data.fillna(method='bfill', inplace=True)
x1=np.array([data.OverallQual])
x2=np.array([data.YearBuilt])
x3=np.array([data.TotalBsmtSF])
x4=np.array([data.GrLivArea])
x5=np.array([data.FullBath])
x6=np.array([data.GarageCars])
X=np.concatenate((x1,x2), axis=0)
X=np.concatenate((X,x3), axis=0)
X=np.concatenate((X,x4), axis=0)
X=np.concatenate((X,x5), axis=0)
X=np.concatenate((X,x6), axis=0)
print('done')
```

```python
pred=model.predict(X.T)
```

```python
pred=pred.reshape((pred.shape[0],))
```

```python
ID=[i for i in range(1461,1461+pred.shape[0])]
```

```python
C={'Id':ID,'SalePrice':pred}
```

```python
df= pd.DataFrame(C,columns=['Id','SalePrice'])
```
