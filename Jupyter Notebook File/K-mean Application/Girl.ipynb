```python
img= img.reshape((img.shape[0]*img.shape[1],3))
```

```python
# s lng cluster = 5
clus=5
m,n=img.shape
Y=np.full((m,clus),0)
```

```python
def dist(a,b):
    return np.sqrt(np.sum((a-b)**2))
```

```python
# test dist function
dist(np.array([0,0]),np.array([0,100]))
```

```python
def kmean_init_centroid():
    ans=np.random.randint(0,255,clus*3)
    ans= ans.reshape((clus,3))
    return ans
```

```python
center=kmean_init_centroid()
```

```python
def kmean_assign_label(M,X):
    m,n  =X.shape
    m1,n1=M.shape
    for i in range(m):
        Min=1000000
        indexI=-1
        indexJ=-1
        for j in range(m1):
            if dist(X[i],M[j])<Min:
                Min=dist(X[i],M[j])
                indexI=i
                indexJ=j
        Y[indexI]=0
        Y[indexI,indexJ]=1
```

```python
def kmean_update_centroid(M,X):
    m,n  = X.shape
    m1,n1= M.shape
    for i in range(m1):
        tmp= np.array([Y[:,i]])
        tmp=tmp.T
        M[i]=np.sum(tmp*X,axis=0)/sum(tmp)
```

```python
iterr=100
while(iterr>=0):
    kmean_assign_label(center,img)
    print(np.sum(Y,axis=0))
    kmean_update_centroid(center,img)
    print(center)
    print(iterr)
    print("===========================================================")
    iterr-=1
```
