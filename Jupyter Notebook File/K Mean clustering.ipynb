```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
X= np.random.randint(70,100,50)
Y= np.random.randint(80,120,50)
X= np.concatenate((X,np.random.randint(0,20,50)),axis=0)
Y= np.concatenate((Y,np.random.randint(2,50,50)),axis=0)
```

```python
X=np.array([X])
Y=np.array([Y])
```

```python
X=np.concatenate((X,Y),axis=0)
```

```python
X=X.T
```

```python
Y=np.zeros(X.shape)
```

```python
def dist(a,b):
    ans=0
    for i,j in zip(a,b):
        ans+=(i-j)**2
    return ans
```

```python
a=[0,0,0]
b=[100,0,100]
dist(a,b)
```

```python
# kmean init center
M=np.array(
    [[15,120],
    [100,60]]
)
```

```python
#kmean assign label
def find_centroid(X,M):
    m,n=X.shape
    for i in range(m):
        if(dist(X[i],M[0])<dist(X[i],M[1])):
            Y[i][0]=1
            Y[i][1]=0
        else:
            Y[i][1]=1
            Y[i][0]=0
```

```python
#kmean update center
def moving_centroid(X,M):
    m,n= M.shape
    tmp=[]
    for i in range(m):
        Y1=np.array([Y[:,i]])
        Y1=Y1.T
        temp=sum(Y1*X)
        temp=temp/sum(Y1)
        M[i]=temp
```

```python
#learning
iter_n=1000
while(iter_n>=0):
    find_centroid(X,M)
    moving_centroid(X,M)
    print('iter_n= {}'.format(1000-iter_n+1))
    print('M = {}'.format(M))
    iter_n-=1
```
