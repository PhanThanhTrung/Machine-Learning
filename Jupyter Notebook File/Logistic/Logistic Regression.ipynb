```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data= pd.read_csv('data.csv')
data.head(3)
```

```python
x1=np.array(data.iloc[:,0])
x2=np.array(data.iloc[:,1])
X =np.concatenate((x1.reshape((1,x1.shape[0])),x2.reshape((1,x2.shape[0]))),axis=0)
one= np.ones_like(x1)
X =np.concatenate((X,one.reshape((1,one.shape[0]))),axis=0)

y=np.array(data.iloc[:,2])
```

```python
X.shape
```

```python
#visualizing data
for i in range(x1.shape[0]):
    if y[i]==1:
        plt.plot(x1[i],x2[i],'ro')
    else:
        plt.plot(x1[i],x2[i],'bo')
plt.show()
```

```python
def init_weight():
    return np.random.rand(1,X.shape[0])
def predict(w,x):
    return 1/(1+np.exp(-np.sum(w*x)))
```

```python
def train_weight(w,X,y,lr,epochs):
    while(epochs>=0):
        mispoint=[]
        print('epochs = {}   || w= {} '.format(epochs,w))
        m,n= X.shape
        for i in range(n):
            xi=X[:,i]
            yi=y[i]
            zi=predict(w,xi)
            if zi!=yi:
                mispoint.append(xi)
            w= w + lr*(yi-zi)*xi
        print('mispoint = {}'.format(len(mispoint)))
        print('==========================================================')
        epochs-=1
    return w
```

```python
w= init_weight()
```

```python
x=[i for i in range(30,100)]
y1=[(-w[0][0]*i - w[0][2])/w[0][1] for i in x]
plt.plot(x,y1)
plt.show()
```

```python
w=train_weight(w,X,y,0.1,10000)
```

```python
for i in range(x1.shape[0]):
    if y[i]==1:
        plt.plot(x1[i],x2[i],'ro')
    else:
        plt.plot(x1[i],x2[i],'bo')
x=[i for i in range(30,100)]
y1=[(-w[0][0]*i - w[0][2])/w[0][1] for i in x]
plt.plot(x,y1)
plt.show()
```
