```python
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
```

## Sinh Dữ Liệu

```python
def test_y(x):
    return 4*x+5+random.uniform(0.1,0.2)
X = np.array([list(range(-50,50,2))])
Y = np.array([test_y(each) for each in X])
```

```python
X=X.T
Y=Y.T
```

```python
def cost(m,c):
    return sum(((Y-m*X-c)**2))/(2*len(X))
```

## Học

```python
m=5
c=6
lr=0.001
epochs=10000
n=len(Y)
for i in range(epochs):
    print('Epochs: ', i+1)
    print('MSE: ', cost(m,c))
    gr_m = (1/n)*sum((m*X+c-Y)*X)
    gr_c = (1/n)*sum((m*X+c-Y))
    m = m - lr*gr_m
    c = c - lr*gr_c
    print('w1={} || w2={} '.format(m,c))
    print('================================')
#print(m,c)
```
